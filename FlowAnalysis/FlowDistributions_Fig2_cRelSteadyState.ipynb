```python
cRel_1_func = interp1d(cRel_1h_4_x, cRel_1h_4_x)
cRel_3_func = interp1d(cRel_3h_x, cRel_3h_x)
cRel_6_func = interp1d(cRel_6h_x, cRel_6h_x)
cRel_Mouse2_func = interp1d(cRel_Mouse2_x, cRel_Mouse2_x)
cRel_Mouse3_func = interp1d(cRel_Mouse3_x, cRel_Mouse3_x)
cRel_Mouse4_func = interp1d(cRel_Mouse4_x, cRel_Mouse4_x)
common_x = np.linspace(1000, 10000, 90000)
cRel_1_y_interp = cRel_1_func(common_x)
cRel_3_y_interp = cRel_3_func(common_x)
cRel_6_y_interp = cRel_6_func(common_x)
cRel_Mouse2_y_interp = cRel_Mouse2_func(common_x)
cRel_Mouse3_y_interp = cRel_Mouse3_func(common_x)
cRel_Mouse4_y_interp = cRel_Mouse4_func(common_x)
jsd_1to3 = distance.jensenshannon(cRel_1_y_interp, cRel_3_y_interp, base=2)
jsd_3to6 = distance.jensenshannon(cRel_3_y_interp, cRel_6_y_interp, base=2)
jsd_1to6 = distance.jensenshannon(cRel_1_y_interp, cRel_6_y_interp, base=2)
jsd_2to3 = distance.jensenshannon(cRel_Mouse2_y_interp, cRel_Mouse3_y_interp, base=2)
jsd_2to4 = distance.jensenshannon(cRel_Mouse2_y_interp, cRel_Mouse4_y_interp, base=2)
jsd_3to4 = distance.jensenshannon(cRel_Mouse3_y_interp, cRel_Mouse4_y_interp, base=2)
jsd_1toM2 = distance.jensenshannon(cRel_1_y_interp, cRel_Mouse2_y_interp, base=2)
jsd_1toM3 = distance.jensenshannon(cRel_1_y_interp, cRel_Mouse3_y_interp, base=2)
jsd_1toM4 = distance.jensenshannon(cRel_1_y_interp, cRel_Mouse4_y_interp, base=2)
print("JSD between 1h and 6h = ", jsd_1to6)
print("JSD between 1h and 3h = ", jsd_1to3)
print("JSD between 3h and 6h = ", jsd_3to6)
print("JSD between M1 and M2 = ", jsd_1toM2)
print("JSD between M1 and M3 = ", jsd_1toM3)
print("JSD between M1 and M4 = ", jsd_1toM4)
print("JSD between M2 and M3 = ", jsd_2to3)
print("JSD between M2 and M4 = ", jsd_2to4)
print("JSD between M3 and M4 = ", jsd_3to4)
```

```python
T1_4_Mean = np.mean(cRelFluor_1h_FL4)
T1_3_Mean = np.mean(cRelFluor_1h_FL3)
T1_2_Mean = np.mean(cRelFluor_1h_FL2)
T3_Mean = np.mean(cRelFluor_3h)
T6_Mean = np.mean(cRelFluor_6h)
M2_Mean = np.mean(cRelFluor_Mouse2)
M3_Mean = np.mean(cRelFluor_Mouse3)
M4_Mean = np.mean(cRelFluor_Mouse4)
print("Mean statistics - ")
print("1h = ", T1_4_Mean, " ; ", T1_3_Mean, " ; ", T1_2_Mean, " ; \n3h = ", T3_Mean, " ; \n6h = ", T6_Mean)
print("Average over 6h = ", np.mean([T1_4_Mean, T3_Mean, T6_Mean]))
print("Spread over 6h = ", np.std([T1_4_Mean, T3_Mean, T6_Mean]))
print("M1 = ", T1_4_Mean, " ; \nM2 = ", M2_Mean, " ; \nM3 = ", M3_Mean, " ; \nM4 = ", M4_Mean)
print("Average over 4 mice = ", np.mean([T1_4_Mean, M2_Mean, M3_Mean, M4_Mean]))
print("Spread over 4 mice = ", np.std([T1_4_Mean, M2_Mean, M3_Mean, M4_Mean]))

T1_4_Var = ss.variation(cRelFluor_1h_FL4)
T1_3_Var = ss.variation(cRelFluor_1h_FL3)
T1_2_Var = ss.variation(cRelFluor_1h_FL2)
T3_Var = ss.variation(cRelFluor_3h)
T6_Var = ss.variation(cRelFluor_6h)
M2_Var = ss.variation(cRelFluor_Mouse2)
M3_Var = ss.variation(cRelFluor_Mouse3)
M4_Var = ss.variation(cRelFluor_Mouse4)
print("\nCV statistics - ")
print("1h = ", T1_4_Var, " ; ", T1_3_Var, " ; ", T1_2_Var, " ; \n3h = ", T3_Var, " ; \n6h = ", T6_Var)
print("Average over 6h = ", np.mean([T1_4_Var, T3_Var, T6_Var]))
print("Spread over 6h = ", np.std([T1_4_Var, T3_Var, T6_Var]))
print("M1 = ", T1_4_Var, " ; \nM2 = ", M2_Var, " ; \nM3 = ", M3_Var, " ; \nM4 = ", M4_Var)
print("Average over 4 mice = ", np.mean([T1_4_Var, M2_Var, M3_Var, M4_Var]))
print("Spread over 4 mice = ", np.std([T1_4_Var, M2_Var, M3_Var, M4_Var]))

T1_4_Skew = ss.skew(cRelFluor_1h_FL4)
T1_3_Skew = ss.skew(cRelFluor_1h_FL3)
T1_2_Skew = ss.skew(cRelFluor_1h_FL2)
T3_Skew = ss.skew(cRelFluor_3h)
T6_Skew = ss.skew(cRelFluor_6h)
M2_Skew = ss.skew(cRelFluor_Mouse2)
M3_Skew = ss.skew(cRelFluor_Mouse3)
M4_Skew = ss.skew(cRelFluor_Mouse4)
print("\nSkew statistics - ")
print("1h = ", T1_4_Skew, " ; ", T1_3_Skew, " ; ", T1_2_Skew, " ; \n3h = ", T3_Skew, " ; \n6h = ", T6_Skew)
print("Average over 6h = ", np.mean([T1_4_Skew, T3_Skew, T6_Skew]))
print("Spread over 6h = ", np.std([T1_4_Skew, T3_Skew, T6_Skew]))
print("M1 = ", T1_4_Skew, " ; \nM2 = ", M2_Skew, " ; \nM3 = ", M3_Skew, " ; \nM4 = ", M4_Skew)
print("Average over 4 mice = ", np.mean([T1_4_Skew, M2_Skew, M3_Skew, M4_Skew]))
print("Spread over 4 mice = ", np.std([T1_4_Skew, M2_Skew, M3_Skew, M4_Skew]))

T1_4_Kurt = ss.kurtosis(cRelFluor_1h_FL4)
T1_3_Kurt = ss.kurtosis(cRelFluor_1h_FL3)
T1_2_Kurt = ss.kurtosis(cRelFluor_1h_FL2)
T3_Kurt = ss.kurtosis(cRelFluor_3h)
T6_Kurt = ss.kurtosis(cRelFluor_6h)
M2_Kurt = ss.kurtosis(cRelFluor_Mouse2)
M3_Kurt = ss.kurtosis(cRelFluor_Mouse3)
M4_Kurt = ss.kurtosis(cRelFluor_Mouse4)
print("\nKurtosis statistics - ")
print("1h = ", T1_4_Kurt, " ; ", T1_3_Kurt, " ; ", T1_2_Kurt, " ; \n3h = ", T3_Kurt, " ; \n6h = ", T6_Kurt)
print("Average over 6h = ", np.mean([T1_4_Kurt, T3_Kurt, T6_Kurt]))
print("Spread over 6h = ", np.std([T1_4_Kurt, T3_Kurt, T6_Kurt]))
print("M1 = ", T1_4_Kurt, " ; \nM2 = ", M2_Kurt, " ; \nM3 = ", M3_Kurt, " ; \nM4 = ", M4_Kurt)
print("Average over 4 mice = ", np.mean([T1_4_Kurt, M2_Kurt, M3_Kurt, M4_Kurt]))
print("Spread over 4 mice = ", np.std([T1_4_Kurt, M2_Kurt, M3_Kurt, M4_Kurt]))

RelA_Var = ss.variation(RelAFluor)
RelA_Skew = ss.skew(RelAFluor)
RelA_Kurt = ss.kurtosis(RelAFluor)
print("RelA statistics - ")
print("CV = ", RelA_Var)
print("Skew = ", RelA_Skew)
print("Kurtosis = ", RelA_Kurt)
```
